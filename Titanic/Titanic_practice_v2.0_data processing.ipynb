```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
test_result=pd.read_csv("gender_submission.csv")
```

```python
train
```

```python
#Correlation bertween Embarked and Survived
train[["Embarked",'Survived']].groupby(train["Embarked"]).mean().sort_values(by="Survived")
```

```python
#Correlation between Sex and Survived
train["Sex"]=train["Sex"].astype(object).fillna("None")
train[["Sex",'Survived']].groupby(train["Sex"]).mean().sort_values(by="Survived")
```

```python
#Correlation between Pclass and Survived
train["Pclass"]=train["Pclass"].astype(object).fillna("None")
train[['Survived']].groupby(train["Pclass"]).mean().sort_values(by="Survived")
```

```python
#Correlation between Age and Survivied
train['Modified_Age']=train['Age'].map(lambda x: x-x%10)
train["Modified_Age"]=train["Modified_Age"].astype(object).fillna(train["Modified_Age"].mean())

test['Modified_Age']=test['Age'].map(lambda x: x-x%10)
test["Modified_Age"]=test["Modified_Age"].astype(object).fillna(test["Modified_Age"].mean())

train[['Survived']].groupby(train["Modified_Age"]).mean().sort_values(by="Survived")
```

```python
#Correlation between SibSp and Survived
train["SibSp"]=train["SibSp"].astype(object).fillna("None")
train[['Survived']].groupby(train["SibSp"]).mean().sort_values(by="Survived")
```

```python
#Correlation between SibSp and Survived
train["Parch"]=train["Parch"].astype(object).fillna("None")
train[['Survived']].groupby(train["Parch"]).mean().sort_values(by="Survived")
```

```python
#Correlation between SibSp and Survived 
#Modification
train["Modified_Cabin"]=train['Cabin']
train["Modified_Cabin"]=train['Modified_Cabin'].astype(object).fillna("-1")
train["Modified_Cabin"]=train['Modified_Cabin'].map(lambda x: x[0])
train["Modified_Cabin"]=train['Modified_Cabin'].map({'T':0,'-':1, 'A':2,'G':3, "C":4,"F":5,"B":6,"E":7,"D":8})

test["Modified_Cabin"]=test['Cabin']
test["Modified_Cabin"]=test['Modified_Cabin'].astype(object).fillna("-1")
test["Modified_Cabin"]=test['Modified_Cabin'].map(lambda x: x[0])
test["Modified_Cabin"]=test['Modified_Cabin'].map({'T':0,'-':1, 'A':2,'G':3, "C":4,"F":5,"B":6,"E":7,"D":8})


train[['Survived']].groupby(train["Modified_Cabin"]).mean().sort_values(by="Survived")
```

```python
train['Modified_Sex']=train['Sex'].map({"male":0,"female":1})
test['Modified_Sex']=test['Sex'].map({"male":0,"female":1})
```

```python
train["Modified_Embarked"]=train["Embarked"].map({"S":0,"Q":1,"C":2})
train["Modified_Embarked"]=train["Modified_Embarked"].astype(object).fillna(train["Modified_Embarked"].mean())

test["Modified_Embarked"]=test["Embarked"].map({"S":0,"Q":1,"C":2})
test["Modified_Embarked"]=test["Modified_Embarked"].astype(object).fillna(test["Modified_Embarked"].mean())
```

```python
train=train.drop(['Name', 'Ticket', 'PassengerId'], axis=1)
test=test.drop(['Name', 'Ticket', 'PassengerId'], axis=1)

train
```
